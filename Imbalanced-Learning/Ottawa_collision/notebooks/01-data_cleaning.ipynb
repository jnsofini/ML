## Load and Examine Data

We load the dataset from 2014 to 2017

```python
df1 = pd.read_excel('../data/2014collisionsfinal.xls')
df2 = pd.read_excel('../data/2015collisionsfinal.xls')
df3 = pd.read_excel('../data/2016collisionsfinal.xls')
df4 = pd.read_excel('../data/h2017collisionsfinal.xlsx')
```

We can see that there are some inconsitency in the four datasets. The 2017 dataset contains Year, whereas others do not. It also contains Longitude and Latitude which is the same as X and Y in the other datasets. Therefore, we will drop these Year, Longitude, and Latitude attributes from the last dataframe

```python
df4 = df4.drop(['Latitude', 'Longitude', 'Year'], axis =1)
```

Now we emerge all the dataset into one dataframe using pandas.concat

```python
df = pd.concat([df1, df2, df3, df4], sort = False)
```

## Feature Engineering

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name() 
df['Week_day'] = df['Date'].dt.weekday_name
```

```python
# Data size 
df.shape
```

```python
# Drop duplicate rows
df = df.drop_duplicates() 
```

## Save data

```python
df.to_csv('../data/data_2014_to_2017.csv', index = False)
```
